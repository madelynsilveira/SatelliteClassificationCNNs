```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your CSVs
cropped_df = pd.read_csv("initial_cropped.csv")
uncropped_df = pd.read_csv("initial_uncropped.csv")

# Ensure model names match and are aligned
models = cropped_df['Model']
test_acc_cropped = cropped_df['test accuracy']
test_acc_uncropped = uncropped_df['test accuracy']

# Create bar positions
x = np.arange(len(models))
width = 0.35

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x - width/2, test_acc_cropped, width, label='Cropped', color='darkslateblue')
bars2 = ax.bar(x + width/2, test_acc_uncropped, width, label='Uncropped', color='orangered')

# Labeling
ax.set_ylabel('Test Accuracy')
ax.set_title('Test Accuracy by Model (Cropped vs. Uncropped)')
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=45, ha='right')
ax.legend()

# Annotate bars with values
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.annotate(f'{height:.2f}',
                xy=(bar.get_x() + bar.get_width() / 2, height),
                xytext=(0, 3),
                textcoords="offset points",
                ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
